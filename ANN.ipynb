```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_model(x,y,units,epochs,batch_size,verbose=0):
    x=x.reshape((x.shape[0],1))
    y=y.reshape((y.shape[0],1))
    model=keras.Sequential()
    model.add(keras.layers.Dense(units=units,activation='relu'))
    model.add(keras.layers.Dense(units=1))
    model.compile(optimizer='adam',loss='mse')
    model.fit(x,y,epochs=epochs,batch_size=batch_size,verbose=verbose)
    return model
```

```python
def show_model(x,y,model):
    plt.plot(x,y)
    plt.plot(x,model.predict(x),color='red')
    plt.legend(['Actual','Predicted'])
    plt.show()
```

```python
def test_model(x,y,units=1000,epochs=200,batch_size=50,verbose=0):
    show_model(x,y,make_model(x,y,units,epochs,batch_size,verbose))
```

```python
x_vals=np.linspace(0,5,10000)
y_vals=3*np.cos(2*x_vals)+5*np.cos(8*x_vals) - np.sin(0.05*x_vals) + 8*np.cos(1.5*x_vals)
plt.plot(x_vals,y_vals)
plt.show()
```

```python
test_model(x_vals,y_vals,units=10000,epochs=50)
```

```python
test_model(x_vals,y_vals,units=10000,epochs=100)
```
